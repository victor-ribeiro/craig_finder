```python
import torch as T
from torch.optim import Adam
from torch import nn
from model.classifier import Classifier, Epoch, train
import pandas as pd
from pathlib import Path
import warnings
import numpy as np
```

```python
warnings.filterwarnings('ignore')
```

```python
class ClassifierDataset(T.utils.data.Dataset):
    def __init__(self, dataframe, target):
        self.dataframe = dataframe
        self.target = target
        self.features = dataframe.columns.drop(target)

    def __getitem__(self, index):
        item = self.dataframe.iloc[index]
        features = np.float32(item[self.features].to_numpy())
        labels = np.float32(item[self.target])
        return features, labels
    def __len__(self):
        return len(self.dataframe)
```

```python
DATA_ROOT = Path('data', 'employee', 't_Employee.csv')
data = pd.read_csv(DATA_ROOT)
target = 'LeaveOrNot'

data.head()
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data, test_size=.2, random_state=42)

train_data = ClassifierDataset(train_data, target)
train_data = T.utils.data.DataLoader(train_data, batch_size=32, shuffle=True)

test_data = ClassifierDataset(test_data, target)
test_data = T.utils.data.DataLoader(test_data, batch_size=32, shuffle=True)
```
